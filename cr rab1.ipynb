```python
l1 = list(range(1,11))
print(l1)
def func2(l1):
    ll1 = list()
    for i in range(len(l1)):
        if (i + 1) %5 == 0:
            ll1.append(l1[i] * 2)
        elif (i + 1) %3 == 0:
            continue  
        elif (i + 1) % 5 != 0 or (i + 1) % 3 != 0:
            ll1.append(l1[i])
    return ll1
print('Итог:', func2(l1))
```
