```python
n = int(input('Введите число:'))
l = [1, 2, 3, 4, 5, 6, 7, 8]
def func1(l, n):
 temp = list()
 for i in l:
  if i % n == 0:
   temp.append(i)
   return temp
print('Выходной список:', func1(l, n))
```

```python
x = list(range(1, 13))
print('Старт:', x)
def func4(x):
 i = 0
 while (i + 2) <= len(x):
  x[i], x[i + 2] = x[i + 2], x[i]
  i += 3
  return x
print('Вывод:', func4(x))
```

```python
s1 = [1, 2, 5, 6, 7, 9]
s2 = [6, 7, 4, 6, 1, -5]
def func5(s1, s2):
 s3 = list()
 i = 0
 while i < len(s1 + s2)/2:
  s3.append(s1[i] - s2[i])
  i += 1
 return s3
print('Выходные данные:', func5(s1, s2))
```

```python
N = [-4,14, 16, 1, 0, -36, 1, -7, 22, 4, 45]
def func7(N):
 new_N = list()
 for i in N:
  if i > 0:
   new_N.append(i)
   new_N.sort()
   return new_N
print('Вывод:', func7(N))
```
